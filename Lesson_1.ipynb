## Задача 1
Подбрасываются три игральных кубика. Оцените вероятность того, что сумма значений этих кубиков не будет превышать 6.

Подсказка: оценкой вероятности может служить, например, относительная частота при большом числе опытов. Попробуйте реализовать наибольшее число опытов, которое позволяют ваш компьютер и/или ваше терпение.

========================================================================================================================

Попробую сначала пойти путем логики, а потом подтвердить это эмпирически))).
С точки зрения оценки вероятности, ее можно оценить как частное возможных исходов, при которых сумма выпавших сторон кубика равна 5 или менее, и количество всех возможных исходов.

Количество всех возможных исходов определить достаточно просто, поскольку у каждого кубика 6 сторон, то количество всех возможных вариантов исходов можно определить перемножением мощности множеств исходов каждого кубика 6х6х6=216. При этом нужно понимать, что исходы (1,1,1), (1,1,1), (1,1,1) - это три разных исхода, характерных для каждого отдельного кубика. Другими словами если мы можем однозначно определить где какой кубик,например, они разного цвета, то каждый исход является комбинацией исходов для каждого кубика. 

Для того чтобы определить количество положительных для нас исходов, достаточно будет определить количество всех возможных комбинаций и умножить их на 3, поскольку для каждого кубика комбинации одинаковы  

(1,1,1)  
(1,1,2)   
(1,1,3)   
(1,2,1)  
(1,2,2)  
(1,3,1)   
Итого исходов для одного кубика 6, поскольку кубиков 3, то всего возможных исходов 18, таким образом итоговая вероятность = 



```python
print(f"Итоговая вероятность {18/216:%}")
```

Посмотрим насколько функция будет отличаться от того что определили логически))

```python
def three_cubes(n):
    positive = 0
    total = 0
    while total != n:
        first_cube = np.random.randint(1, 6)
        second_cube = np.random.randint(1, 6)
        third_cube = np.random.randint(1, 6)
        if (first_cube + second_cube + third_cube) < 6:
            positive += 1
        total += 1
    return positive/total
```

```python
list_of_probability = []
for i in range (2000, 22000, 2000):
    list_of_probability.append(three_cubes(i))
```

```python
list_of_probability
```

Ну в целом эмпирически получается что вероятность определена где-то на интервале от 7% до 8.5%.

## Задача 2
Имеется колода в 52 карты. Найти число возможностей вытянуть из неё 4 карты так, чтобы среди них был хотя бы один туз.

==============================================================================================================

Во-первых в задаче дано неопределенное условие. Вытянуть из колоды 4 карты, `чтобы был хотя бы один туз`. Таким образом нужно понимать что могут быть следующие ситуации:  
1. Не вытянуто ни одного туза
2. Вытянут один туз
3. Вытянуто два туза
4. Вытянуто три туза
5. Вытянуто 4 туза    

Нам необходимо оценить сумму количества возможных исходов со второго по пятый пункт  
Количество исходов с помощью которых можно вытянуть один туз можно определить как число сочетаний вытянуть один туз из 4 и умноженное на число сочетаний вытянуть три карты из оставшихся 48  
Количетсво из пункта 3 и 4 определяется точно таким же образом. В пункте 5 количество возможных исходов равно 1, поскольку такая ситуация когда вытянуто сразу 4 туза, при 4 возможных картах по умолчанию 1.  
Далее полученные количетсва необходимо сложить и разделить на общее число возможных исходов
Общая число исходов можно определить как общее число сочетаний из 52 по 4 элементам.

```python
def combination(n, k):
    return math.factorial(n) / (math.factorial(k) * math.factorial(n-k))
```

```python
one_ace = combination(4, 1) * combination(48, 3)
two_aces = combination(4, 2) * combination(48, 2)
three_aces = combination(4, 3) * combination(48, 1)
total_combinations = combination(52, 4)
```

```python
print(f"Итоговая вероятность равна = '{(one_ace + two_aces + three_aces + 1)/total_combinations:%}'")
```

## Задача 4

Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?

=========================================================================================================================

События являются независимыми, если наступление события В не зависит от наступления или ненаступления события А  
Если рассматривать пример из задачи, то например событие В может произойти когда кубик упадет сторонами 4,5,6. При этом появление четного числа происходит в двух случаях из трех. При этом появление четного числа может происходить и когда кубик упадет стороной 2. Другими словами появление четного числа на кубике не означает что число точно больше или меньше 3 и наоборот если выпало число больше трех это не означает что оно четное. Таким образом описываемые события являются независимыми по отношению друг к другу

## Задача 5

Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.

Вопрос: какова вероятность, что вы действительно больны ей?

=====================================================================================================================

В данном случае нужно исходить из отношения при котором в числителе у нас будет реальный положительный тест или True Positive. В данном случае его можно определить как 0.1% х 99%  
В знаменателе должна быть оценка полной вероятности исходов того что тест показал положительный результат:  
1. True positive - я болен и тест это показал 0.1% х 99%
2. False positive - я не болен, но тест это показал 99,9% х 1%

таким образом вероятность что я действительно болен (тьфу-тьфу-тьфу) этой болезнью оценивается на уровне

```python
print(f"Итоговая вероятность равна = '{(0.001 * 0.99) / ((0.001 * 0.99) + (0.999 * 0.01)):%}'")
```
